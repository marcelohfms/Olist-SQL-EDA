# Análise utilizando SQL - Respondendo a perguntas de negócio

## 1- Qual a distribuição de consumidores por estado?

```python
query = "SELECT COUNT(customer_id) as QuantidadeClientes, customer_state AS Estado \
         FROM customers \
         GROUP BY Estado"
psql.sqldf(query)
```

```python
# Visão ordenada
query = "SELECT COUNT(customer_id) as QuantidadeClientes, customer_state AS Estado \
         FROM customers \
         GROUP BY Estado \
         ORDER BY QuantidadeClientes DESC"
psql.sqldf(query)
```

## 2- Quantos pedidos fizeram os 10 clientes que mais compraram?

Para responder essa pergunta, é preciso utilizar dados das tabelas _customers_ e _orders_ juntas. Portanto, faremos um JOIN.

```python
query = "SELECT COUNT(t1.order_id) AS QuantidadePedidos, t2.customer_unique_id AS IdCliente \
         FROM orders t1\
         INNER JOIN customers t2 ON t1.customer_id = t2.customer_id \
         GROUP BY IdCliente \
         ORDER BY QuantidadePedidos DESC \
         LIMIT 10"
psql.sqldf(query)
```

## 3- Quais foram as 3 categorias de itens mais compradas em 2017?

Para responder essa pergunta, precisamos de informações da tabela _orders_, _products_ e também da tabela _order_items_, vamos então fazer dois joins.

```python
# Convertendo data do pedido para formato datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders.info()
```

```python
# obtendo uma tabela dos itens comprados em 2017
query = "SELECT t1.order_id AS IdPedido, t1.order_item_id AS IdItemPedido, t1.product_id AS IdProduto \
         FROM order_items t1 \
         INNER JOIN orders t2 ON t1.order_id = t2.order_id \
         WHERE strftime('%Y', t2.order_purchase_timestamp) = '2017'"

orders_2017 = psql.sqldf(query)
```

```python
# Identificando o tipo de produto
query = "SELECT t1.IdPedido, t1.IdItemPedido, t1.IdProduto, t2.product_category_name AS CategoriaProduto\
         FROM orders_2017 t1 \
         INNER JOIN products t2 \
         ON t1.IdProduto = t2.product_id"
orders_2017_categoria = psql.sqldf(query)
orders_2017_categoria
```

```python
# Contagem por Categoria de Produto
query = "SELECT COUNT(IdPedido) AS QuantidadeItens, CategoriaProduto \
         FROM orders_2017_categoria \
         GROUP BY CategoriaProduto \
         ORDER BY QuantidadeItens DESC \
         LIMIT 3"
psql.sqldf(query)
```

```python
# Poderíamos ter realizado o processo todo com uma única query utilizando de subqueries, porém seria muito menos legível
query = "SELECT COUNT(IdPedido) AS QuantidadeItens, CategoriaProduto \
         FROM (SELECT t1.IdPedido, t1.IdItemPedido, t1.IdProduto, t2.product_category_name AS CategoriaProduto\
               FROM (SELECT t1.order_id AS IdPedido, t1.order_item_id AS IdItemPedido, t1.product_id AS IdProduto \
                     FROM order_items t1 \
                     INNER JOIN orders t2 ON t1.order_id = t2.order_id \
                     WHERE strftime('%Y', t2.order_purchase_timestamp) = '2017') t1 \
               INNER JOIN products t2 \
               ON t1.IdProduto = t2.product_id) \
         GROUP BY CategoriaProduto \
         ORDER BY QuantidadeItens DESC \
         LIMIT 3"
psql.sqldf(query)
```

## 4- Quantos e quais vendedores tiveram um ticket médio por pedido maior que 150 em 2018?

Para responder essa pergunta, precisamos de informações da tabela _orders_, _sellers_ e também da tabela order_items, então vamos fazer dois JOINs. Além disso, é preciso filtrar um agrupamento, então será necessário usar a cláusula HAVING.

```python
# obtendo uma tabela dos pedidos feitos em 2018
query = "SELECT t1.order_id AS IdPedido, t1.seller_id AS Vendedor, t1.price AS Preco \
         FROM order_items t1 \
         INNER JOIN orders t2 ON t1.order_id = t2.order_id \
         WHERE strftime('%Y', t2.order_purchase_timestamp) = '2018'"

orders_2018 = psql.sqldf(query)
orders_2018
```

```python
# Agrupando o valor do ticket por pedido
query = "SELECT IdPedido, Vendedor, SUM(Preco) AS TicketVenda \
         FROM orders_2018 \
         GROUP BY IdPedido"
orders_2018_ticket = psql.sqldf(query)
orders_2018_ticket
```

```python
# Obtendo o Ticket médio por vendedor e filtrando > 150
query = "SELECT Vendedor, AVG(TicketVenda) AS TicketMedio \
         FROM orders_2018_ticket \
         GROUP BY Vendedor \
         HAVING TicketMedio > 150 \
         ORDER BY TicketMedio DESC"

psql.sqldf(query)
```

## 5- Quantos e que proporção do total de vendedores tiveram ticket médio por pedido maior que 250 em 2018?

Vamos utilizar uma CTE (Common Table Expression) modificada da query anterior, e depois fazer uma outra query com SELECT para encontrar essa informação

```python
# Obtendo
query = """
WITH ticket_table AS (
    SELECT Vendedor, AVG(TicketVenda) AS TicketMedio
    FROM orders_2018_ticket
    GROUP BY Vendedor
)

SELECT COUNT(Vendedor) AS QuantidadeVendedores,
       ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM ticket_table), 2) AS Proporcao
FROM ticket_table
WHERE TicketMedio > 250;
"""

psql.sqldf(query)
```

**OBS**: Da forma que a consultas foram implementadas, **apenas os vendedores que tiveram ao menos uma venda em 2018 foram considerados para o cálculo da Proporção.**

## 6 - Qual o preço médio do frete para os pedidos com 3 ou mais itens?

A especificação do dataset diz que quando há mais de um item por pedido, o preço do frete é dividido entre eles. Logo é preciso ajustar o valor da coluna _freight_value_ pela quantidade de itens no pedido.

```python
query = "SELECT AVG(PrecoFrete) AS PrecoMedioFrete \
         FROM (SELECT MAX(order_item_id) AS QuantidadeItens, order_id AS IdPedido, freight_value*MAX(order_item_id) AS PrecoFrete \
               FROM order_items \
               GROUP BY IdPedido \
               HAVING QuantidadeItens >= 3) \
        "
psql.sqldf(query)
```

## 7 - Qual o preço médio do frete para cada quantidade de itens acima de 3 por pedido?

```python
query = "SELECT QuantidadeItens, AVG(PrecoFrete) AS PrecoMedioFrete \
         FROM (SELECT MAX(order_item_id) AS QuantidadeItens, order_id AS IdPedido, freight_value*MAX(order_item_id) AS PrecoFrete \
               FROM order_items \
               GROUP BY IdPedido \
               HAVING QuantidadeItens >= 3) \
         GROUP BY QuantidadeItens \
        "
psql.sqldf(query)
```

## 8 - Qual foi a evolução da quantidade de pedidos mês a mês? Quantos pedidos cada mês teve a mais ou a menos que o mês anterior?

```python
query = """
SELECT COUNT(order_id) AS QuantidadePedidos, strftime('%m', order_purchase_timestamp) AS Mes, strftime('%Y', order_purchase_timestamp) AS Ano
FROM orders
GROUP BY Mes, Ano
ORDER BY Ano, Mes
"""

orders_by_month = psql.sqldf(query)
orders_by_month
```

```python
# Ajustando colunas mês e ano para serem apenas uma
query = """
SELECT QuantidadePedidos,
       strftime('%Y-%m-01', Ano || '-' || CASE WHEN LENGTH(Mes) = 1 THEN '0' || Mes ELSE Mes END || '-01') AS MesAno
FROM orders_by_month
"""
orders_by_month = psql.sqldf(query)
```

```python
query = """
SELECT MesAno, QuantidadePedidos, QuantidadePedidos-LAG(QuantidadePedidos, 1) OVER() AS DeltaMesAnterior
FROM orders_by_month
"""

psql.sqldf(query)
```

## 9 - Quais foram os 5 clientes que mais gastaram? Qual foi o ticket médio por item deles?

```python
query = """ 
WITH customer_orders AS (
    SELECT t1.order_id as IdPedido, t1.price as Valor, t2.customer_id as IdCliente, t3.customer_unique_id AS IdUnicoCliente
    FROM order_items t1
    INNER JOIN orders t2
    ON t1.order_id = t2.order_id
    INNER JOIN customers t3
    ON t2.customer_id = t3.customer_id
)

SELECT IdUnicoCliente, RANK() OVER (ORDER BY SUM(Valor) DESC) AS RankCliente, SUM(Valor) AS TotalGasto,
        AVG(Valor) AS TicketMedio
FROM customer_orders
GROUP BY IdUnicoCliente
LIMIT 5
"""

psql.sqldf(query)
```

## 10 - Quais são os 5 vendedores com maior número de vendas, e qual o total de vendas deles em reais?

```python
query = """ 
WITH seller_orders AS (
    SELECT order_id as IdPedido, price as Valor, seller_id as IdVendedor
    FROM order_items
)

SELECT IdVendedor, RANK() OVER (ORDER BY COUNT(DISTINCT IdPedido) DESC) AS RankVendedor, COUNT(DISTINCT IdPedido) AS NumPedidos,
        SUM(Valor) AS TotalVendido
FROM seller_orders
GROUP BY IdVendedor
LIMIT 5
"""

psql.sqldf(query)
```
